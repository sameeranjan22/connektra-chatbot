```python
import pandas as pd
import ast
```

```python
df1=pd.read_csv("/content/Connectors.csv")
```

```python
con_Trig=[]
for i in df1["Trigger Variables"]:
  result_list = ast.literal_eval(i)
  con_Trig.extend(result_list)
print(con_Trig)
con_Act=[]
for i in df1["Action Variables"]:
  result_list = ast.literal_eval(i)
  con_Act.extend(result_list)
print(con_Trig)
```

```python
Triggers=list(set(con_Trig))
Actions=list(set(con_Act))
print(Actions)
print(Triggers)
```

```python
apis = ["salesforce", "hubspot", "slack", "Google Sheets", "gmail", "bitrix24", "calendly", "mailchimp", "stripe", "airtable", "twilio", "razorpay", "figma", "intercom", "wordpress", "zoom", "typeform", "jira", "bamboohr", "zendesk", "monday"]
```

```python
words=apis+Triggers+Actions
```

```python
words
```

```python
from spellchecker import SpellChecker

spell = SpellChecker(distance=2)

# find those words that may be misspelled
spell.word_frequency.load_words(words)
misspelled = spell.unknown(['google-sheets', 'calendle', 'sal@dforce','hulspot','Slamk'])

for word in misspelled:
    # Get the one `most likely` answer
    print(spell.correction(word))

    # Get a list of `likely` options
    print(spell.candidates(word))
```
